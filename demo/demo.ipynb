```python
import matplotlib.pyplot as plt
from matplotlib import colors as plt_colors
import subprocess
```

```python
def show_data(rounding=-1):
    contents = open("./artifacts/output.txt").read()
    y_layers = contents.split("\n")[:-1]

    # convert to floats
    data = []
    for y in y_layers:
        row = []
        for x in y.split(","):
            row.append(float(x))
        data.append(row)

    if (rounding != -1):
        fig, axes = plt.subplots(1, 3, figsize=(12, 8))


        colors = [
            [34, 84, 156],
            [47, 105, 190],
            [92, 144, 220],
            [225, 210, 159],
            [68, 151, 92]
        ]

        cols = [[val / 255 for val in col] for col in colors]

        rounded = [[round(col * rounding)/rounding for col in row] for row in data]
        cmap = plt_colors.ListedColormap(cols)
        ranges = [-50, -10, -1, 0.5, 2, 50]
        norm = plt_colors.BoundaryNorm(ranges, cmap.N)

        axes[0].imshow(data)
        axes[1].imshow(rounded, interpolation="none")
        axes[2].imshow(data, interpolation="none", cmap=cmap, norm=norm)

        axes[0].set_title("Generated noise")
        axes[1].set_title("Clipping")
        axes[2].set_title("Color")



    else:
        fig, ax = plt.subplots(1)
        ax.imshow(data)
```

## Compile

```python
subprocess.run(["mkdir", "-p", "./artifacts"])
subprocess.run(["g++", "-O3", "demo.cpp", "../src/perlin.cpp", "-o", "./artifacts/demo"])
```

## Run

### Multi-layer noise
This example layers multiple frequencies with different amplitudes, which can be set by changing the $ampltiudes$ and $frequencies$ arrays.

For this demo, frequencies below $\sim 0.01$ and frequencies above $\sim 0.2$ will be difficult to see.

```python
amplitudes = [10, 5, 2]
frequencies = [0.02, 0.1, 0.2]

run_args = ["./artifacts/demo"]
for i in range(min(len(amplitudes), len(frequencies))):
    run_args.append(f"{amplitudes[i]}")
    run_args.append(f"{frequencies[i]}")

subprocess.run(run_args)
show_data(0.3)
```

### Single-layer noise
This example layers a single frequency.

```python
subprocess.run(["./artifacts/demo"])
show_data(5)
```
